Q1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import scipy.stats as stats

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
  """
  Performs a one-sample Z-test.

  Args:
    sample_mean: The mean of the sample.
    population_mean: The hypothesized population mean.
    population_std: The population standard deviation.
    sample_size: The size of the sample.
    alpha: Significance level (default: 0.05).

  Returns:
    z_score: The calculated Z-score.
    p_value: The p-value of the test.
  """

  z_score = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))
  p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

  return z_score, p_value

# Example usage:
sample_mean = 102  # Sample mean
population_mean = 100  # Population mean
population_std = 15  # Population standard deviation
sample_size = 50  # Sample size
alpha = 0.05  # Significance level

z_score, p_value = z_test(sample_mean, population_mean, population_std, sample_size, alpha)

print(f"Z-score: {z_score}")
print(f"p-value: {p_value}")

if p_value < alpha:
  print("Reject the null hypothesis. There is sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the sample mean is significantly different from the population mean.")
```

Q3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
import numpy as np
from scipy import stats

def one_sample_z_test(sample, population_mean, population_std):
  """
  Performs a one-sample Z-test to compare the sample mean with the population mean.

  Args:
    sample: A list or NumPy array containing the sample data.
    population_mean: The population mean.
    population_std: The population standard deviation.

  Returns:
    z_statistic: The calculated Z-statistic.
    p_value: The p-value associated with the Z-statistic.
  """

  sample_mean = np.mean(sample)
  sample_size = len(sample)
  z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
  p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

  return z_statistic, p_value

# Example usage:
sample_data = [89, 93, 95, 93, 97, 98, 96, 99, 93, 97, 110, 104, 119, 105, 104, 110, 110, 112, 115, 114]
population_mean = 100
population_std = 15

z_stat, p_value = one_sample_z_test(sample_data, population_mean, population_std)

print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation:
alpha = 0.05  # Significance level
if p_value < alpha:
  print("Reject the null hypothesis. Sample mean is significantly different from population mean.")
else:
  print("Fail to reject the null hypothesis. Sample mean is not significantly different from population mean.")
```

Q5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type_1_type_2_errors(mu0, mu1, sigma, alpha, n):
  """
  Visualizes Type I and Type II errors for a one-tailed hypothesis test.

  Args:
    mu0: Mean under the null hypothesis.
    mu1: Mean under the alternative hypothesis.
    sigma: Standard deviation.
    alpha: Significance level (Type I error rate).
    n: Sample size.
  """

  # Calculate standard error
  se = sigma / np.sqrt(n)

  # Calculate critical value
  z_alpha = norm.ppf(1 - alpha)

  # Calculate critical value in terms of sample mean
  x_crit = mu0 + z_alpha * se

  # Calculate z-score for mu1 under the null hypothesis
  z_mu1 = (mu1 - mu0) / se

  # Calculate probability of Type II error (beta)
  beta = norm.cdf(z_alpha - z_mu1)

  # Create x-axis for visualization
  x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)

  # Calculate probability density functions
  null_pdf = norm.pdf(x, mu0, se)
  alt_pdf = norm.pdf(x, mu1, se)

  # Create figure
  fig, ax = plt.subplots(figsize=(10, 6))

  # Plot null and alternative distributions
  ax.plot(x, null_pdf, label='Null Hypothesis')
  ax.plot(x, alt_pdf, label='Alternative Hypothesis')

  # Shade Type I error region
  ax.fill_between(x, null_pdf, where=(x > x_crit), alpha=0.3, color='red', label='Type I Error')

  # Shade Type II error region
  ax.fill_between(x, alt_pdf, where=(x <= x_crit), alpha=0.3, color='blue', label='Type II Error')

  # Add vertical lines for critical value and mu1
  ax.axvline(x_crit, color='black', linestyle='--', label='Critical Value')
  ax.axvline(mu1, color='green', linestyle='--', label='True Mean (mu1)')

  # Add labels and legend
  ax.set_xlabel('Sample Mean')
  ax.set_ylabel('Probability Density')
  ax.set_title('Type I and Type II Errors')
  ax.legend()

  # Display the plot
  plt.show()

  # Print error probabilities
  print(f"Type I Error Rate (alpha): {alpha}")
  print(f"Type II Error Rate (beta): {beta}")
  print(f"Power of the Test (1-beta): {1-beta}")

# Example usage
mu0 = 0  # Mean under null hypothesis
mu1 = 2  # Mean under alternative hypothesis
sigma = 1  # Standard deviation
alpha = 0.05  # Significance level
n = 30  # Sample size

visualize_type_1_type_2_errors(mu0, mu1, sigma, alpha, n)
```

Q6.Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy import stats

def independent_t_test(data1, data2, alpha=0.05):
  """
  Performs an independent two-sample t-test.

  Args:
    data1: The first sample data.
    data2: The second sample data.
    alpha: Significance level (default: 0.05).

  Returns:
    t_statistic: The calculated t-statistic.
    p_value: The p-value associated with the t-statistic.
  """

  t_statistic, p_value = stats.ttest_ind(data1, data2, equal_var=False) 

  print("T-statistic:", t_statistic)
  print("P-value:", p_value)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence of a statistically significant difference between the two groups.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence of a statistically significant difference between the two groups.")

# Example usage:
data1 = [25, 30, 28, 32, 27, 26, 29]
data2 = [35, 40, 38, 42, 37, 36, 39]

independent_t_test(data1, data2)
```

Q8.Simulate data and perform both Z-test and T-test, then compare the results using Python

```python
import numpy as np
from scipy import stats

def simulate_and_test(mu0, mu1, sigma, n, alpha=0.05):
    """
    Simulates data, performs Z-test and T-test, and compares the results.

    Args:
        mu0: Population mean under the null hypothesis.
        mu1: Population mean under the alternative hypothesis.
        sigma: Population standard deviation.
        n: Sample size.
        alpha: Significance level (default: 0.05).

    Returns:
        z_stat: Z-statistic.
        z_pvalue: P-value from Z-test.
        t_stat: T-statistic.
        t_pvalue: P-value from T-test.
    """

    # Simulate data
    data = np.random.normal(loc=mu1, scale=sigma, size=n)

    # Calculate sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation

    # Z-test
    z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))
    z_pvalue = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # Two-tailed test

    # T-test
    t_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))
    t_pvalue = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))  # Two-tailed test

    return z_stat, z_pvalue, t_stat, t_pvalue

# Example usage
mu0 = 50  # Population mean under null hypothesis
mu1 = 55  # Population mean under alternative hypothesis
sigma = 10  # Population standard deviation
n = 30  # Sample size

z_stat, z_pvalue, t_stat, t_pvalue = simulate_and_test(mu0, mu1, sigma, n)

print("Z-statistic:", z_stat)
print("Z-test p-value:", z_pvalue)
print("T-statistic:", t_stat)
print("T-test p-value:", t_pvalue)

# Compare results
if z_pvalue < alpha and t_pvalue < alpha:
    print("Both Z-test and T-test reject the null hypothesis.")
elif z_pvalue < alpha and t_pvalue >= alpha:
    print("Z-test rejects the null hypothesis, but T-test fails to reject.")
elif z_pvalue >= alpha and t_pvalue < alpha:
    print("T-test rejects the null hypothesis, but Z-test fails to reject.")
else:
    print("Both Z-test and T-test fail to reject the null hypothesis.")
```

Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence_level=0.95):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    data: A list or NumPy array containing the sample data.
    confidence_level: The desired confidence level (default: 0.95).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """

  n = len(data)
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof=1)  # Sample standard deviation

  # Calculate the critical value (t-distribution)
  t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

  # Calculate margin of error
  margin_of_error = t_critical * (sample_std / np.sqrt(n))

  # Calculate confidence interval
  lower_bound = sample_mean - margin_of_error
  upper_bound = sample_mean + margin_of_error

  return lower_bound, upper_bound

# Example usage:
data = [25, 30, 28, 32, 27, 26, 29]
confidence_level = 0.95

lower, upper = confidence_interval(data, confidence_level)

print(f"Confidence Interval ({confidence_level*100}%): ({lower:.2f}, {upper:.2f})")
```

Q11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def bayesian_inference(prior_alpha, prior_beta, data):
  """
  Performs Bayesian inference for a Bernoulli process.

  Args:
    prior_alpha: Alpha parameter of the Beta prior distribution.
    prior_beta: Beta parameter of the Beta prior distribution.
    data: A list of 0s and 1s representing the observed data.

  Returns:
    posterior_alpha: Alpha parameter of the posterior Beta distribution.
    posterior_beta: Beta parameter of the posterior Beta distribution.
  """

  num_successes = sum(data)
  num_failures = len(data) - num_successes

  posterior_alpha = prior_alpha + num_successes
  posterior_beta = prior_beta + num_failures

  return posterior_alpha, posterior_beta

# Example usage:
prior_alpha = 1
prior_beta = 1  # Uniform prior
data = [1, 0, 1, 1, 0, 1, 1, 1, 0, 0]

posterior_alpha, posterior_beta = bayesian_inference(prior_alpha, prior_beta, data)

# Plot prior and posterior distributions
x = np.linspace(0, 1, 100)
prior_dist = beta.pdf(x, prior_alpha, prior_beta)
posterior_dist = beta.pdf(x, posterior_alpha, posterior_beta)

plt.plot(x, prior_dist, label='Prior')
plt.plot(x, posterior_dist, label='Posterior')
plt.xlabel('Probability of Success (θ)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(f"Posterior Alpha: {posterior_alpha}")
print(f"Posterior Beta: {posterior_beta}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_test(data, var1, var2, alpha=0.05):
  """
  Performs a Chi-square test for independence between two categorical variables.

  Args:
    data: Pandas DataFrame containing the data.
    var1: Name of the first categorical variable.
    var2: Name of the second categorical variable.
    alpha: Significance level (default: 0.05).

  Returns:
    chi2_stat: Chi-square statistic.
    p_value: P-value.
    dof: Degrees of freedom.
    expected_freq: Expected frequencies.
  """

  observed_freq = pd.crosstab(data[var1], data[var2])
  chi2_stat, p_value, dof, expected_freq = chi2_contingency(observed_freq)

  print("Observed Frequencies:\n", observed_freq)
  print("Expected Frequencies:\n", expected_freq)
  print("Chi-Square Statistic:", chi2_stat)
  print("P-value:", p_value)
  print("Degrees of Freedom:", dof)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence of a significant association between", var1, "and", var2)
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence of a significant association between", var1, "and", var2)

# Example Usage
# Sample data (replace with your actual data)
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 
                  'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'SmokingStatus': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 
                         'No', 'No', 'Yes', 'No', 'Yes', 'No']}
df = pd.DataFrame(data)

chi_square_test(df, 'Gender', 'SmokingStatus')
```

Q14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.






```python
import numpy as np
from scipy.stats import chisquare

def goodness_of_fit_test(observed_data, expected_data, alpha=0.05):
  """
  Performs a Chi-Square goodness-of-fit test.

  Args:
    observed_data: A list or NumPy array of observed frequencies.
    expected_data: A list or NumPy array of expected frequencies.
    alpha: Significance level (default: 0.05).

  Returns:
    chi2_stat: The Chi-Square statistic.
    p_value: The p-value.
  """

  chi2_stat, p_value = chisquare(observed_data, f_exp=expected_data)

  print("Chi-Square Statistic:", chi2_stat)
  print("p-value:", p_value)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("The observed data significantly differs from the expected distribution.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no significant evidence that the observed data differs from the expected distribution.")

  return chi2_stat, p_value

# Example usage:
observed_data = np.array([25, 30, 45, 50])  # Observed frequencies
expected_data = np.array([37.5, 37.5, 37.5, 37.5])  # Expected frequencies (e.g., uniform distribution)

goodness_of_fit_test(observed_data, expected_data)
```

Q16.Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

def f_test_for_variance(sample1, sample2, alpha=0.05):
  """
  Performs an F-test to compare the variances of two samples.

  Args:
    sample1: The first sample data.
    sample2: The second sample data.
    alpha: Significance level (default: 0.05).

  Returns:
    f_statistic: The calculated F-statistic.
    p_value: The p-value associated with the F-statistic.
  """

  # Calculate sample variances
  var1 = np.var(sample1, ddof=1)  # Sample variance
  var2 = np.var(sample2, ddof=1)  # Sample variance

  # Determine which sample has larger variance
  if var1 > var2:
    f_statistic = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
  else:
    f_statistic = var2 / var1
    df2 = len(sample1) - 1
    df1 = len(sample2) - 1

  # Calculate p-value
  p_value = 2 * (1 - f.cdf(f_statistic, df1, df2))  # Two-tailed test

  # Print results
  print("F-statistic:", f_statistic)
  print("p-value:", p_value)

  # Interpretation
  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence that the variances of the two populations are significantly different.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence that the variances of the two populations are significantly different.")

# Example usage:
sample1 = [25, 30, 28, 32, 27, 26, 29]
sample2 = [35, 40, 38, 42, 37, 36, 39]

f_test_for_variance(sample1, sample2)
```

Q17.Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

```python
import scipy.stats as stats

def anova_test(groups, alpha=0.05):
  """
  Performs an ANOVA test to compare means between multiple groups.

  Args:
    groups: A list of lists, where each sublist contains the data for one group.
    alpha: Significance level (default: 0.05).

  Returns:
    f_statistic: The calculated F-statistic.
    p_value: The p-value associated with the F-statistic.
  """

  f_statistic, p_value = stats.f_oneway(*groups)

  print("F-statistic:", f_statistic)
  print("p-value:", p_value)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence of statistically significant differences between group means.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence of statistically significant differences between group means.")

# Example usage:
group1 = [10, 12, 15, 18, 20]
group2 = [8, 10, 11, 13, 14]
group3 = [15, 17, 19, 21, 23]

anova_test([group1, group2, group3])
```

Q18.Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def perform_anova_and_plot(groups):
  """
  Performs one-way ANOVA test and creates a box plot to visualize group differences.

  Args:
    groups: A list of lists, where each sublist contains the data for one group.

  Returns:
    f_statistic: The calculated F-statistic.
    p_value: The p-value associated with the F-statistic.
  """

  f_statistic, p_value = stats.f_oneway(*groups)

  # Create a box plot to visualize group differences
  plt.figure(figsize=(8, 5))
  sns.boxplot(data=groups)
  plt.xlabel("Groups")
  plt.ylabel("Values")
  plt.title("Box Plot of Groups")
  plt.show()

  print("F-statistic:", f_statistic)
  print("p-value:", p_value)

  if p_value < 0.05:
    print("Reject the null hypothesis. There is evidence of statistically significant differences between group means.")
  else:
    print("Fail to reject the null hypothesis. There is no evidence of statistically significant differences between group means.")

  return f_statistic, p_value

# Example usage:
group1 = [10, 12, 15, 18, 20]
group2 = [8, 10, 11, 13, 14]
group3 = [15, 17, 19, 21, 23]

f_stat, p_value = perform_anova_and_plot([group1, group2, group3])
```

Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def check_anova_assumptions(groups, alpha=0.05):
  """
  Checks the assumptions of ANOVA: normality, independence, and homogeneity of variances.

  Args:
    groups: A list of lists, where each sublist contains the data for one group.
    alpha: Significance level for normality tests (default: 0.05).

  Returns:
    normality_checks: A list of booleans, True if normality is met for each group.
    homogeneity_check: True if the assumption of equal variances is met, False otherwise.
  """

  normality_checks = []
  for group in groups:
    stat, p_value = stats.shapiro(group)
    normality_checks.append(p_value > alpha)  # Normality is met if p-value > alpha

  # Test for homogeneity of variances (Bartlett's test)
  stat, p_value = stats.bartlett(*groups)
  homogeneity_check = p_value > alpha

  # Visualize data distribution
  sns.boxplot(data=groups)
  plt.xlabel("Groups")
  plt.ylabel("Values")
  plt.title("Boxplot of Groups")
  plt.show()

  return normality_checks, homogeneity_check

# Example usage:
group1 = [10, 12, 15, 18, 20]
group2 = [8, 10, 11, 13, 14]
group3 = [15, 17, 19, 21, 23]

normality_checks, homogeneity_check = check_anova_assumptions([group1, group2, group3])

print("Normality Checks:", normality_checks)
print("Homogeneity of Variances:", homogeneity_check)
```

Q20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

def two_way_anova(data, factor1, factor2, response):
  """
  Performs two-way ANOVA and creates an interaction plot.

  Args:
    data: pandas DataFrame containing the data.
    factor1: Name of the first factor variable.
    factor2: Name of the second factor variable.
    response: Name of the response variable.

  Returns:
    anova_table: ANOVA table as a pandas DataFrame.
  """

  # Create the model
  model = ols(f'{response} ~ C({factor1}) + C({factor2}) + C({factor1})*C({factor2})', data=data).fit() 

  # Perform two-way ANOVA
  anova_table = sm.stats.anova_lm(model, typ=2) 

  # Create an interaction plot
  sns.set_style("whitegrid")
  sns.lineplot(x=factor1, y=response, hue=factor2, data=data)
  plt.title("Interaction Plot")
  plt.xlabel(factor1)
  plt.ylabel(response)
  plt.legend(title=factor2)
  plt.show()

  return anova_table

# Example Usage
data = {'Factor1': ['A', 'A', 'A', 'B', 'B', 'B', 'A', 'A', 'A', 'B', 'B', 'B'],
        'Factor2': ['X', 'X', 'Y', 'X', 'X', 'Y', 'Y', 'Y', 'X', 'Y', 'Y', 'X'],
        'Response': [10, 12, 15, 8, 10, 11, 17, 19, 21, 13, 14, 16]}
df = pd.DataFrame(data)

anova_table = two_way_anova(df, 'Factor1', 'Factor2', 'Response')
print(anova_table)
```

Q21.Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def visualize_f_distribution(df1, df2):
  """
  Visualizes the F-distribution.

  Args:
    df1: Degrees of freedom for the numerator.
    df2: Degrees of freedom for the denominator.
  """

  x = np.linspace(0, 10, 1000)  # Adjust x-range as needed
  pdf = f.pdf(x, df1, df2)

  plt.figure(figsize=(8, 5))
  plt.plot(x, pdf, label=f'F-distribution (df1={df1}, df2={df2})')
  plt.xlabel('F-statistic')
  plt.ylabel('Probability Density')
  plt.title(f'F-distribution')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example usage
df1 = 5  # Degrees of freedom for the numerator
df2 = 10  # Degrees of freedom for the denominator
visualize_f_distribution(df1, df2)
```

Q24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

def chi_square_test_for_variance(sample, population_variance, alpha=0.05):
  """
  Performs a chi-square test for population variance.

  Args:
    sample: A list or NumPy array containing the sample data.
    population_variance: The hypothesized population variance.
    alpha: Significance level (default: 0.05).

  Returns:
    chi_square_stat: The calculated chi-square statistic.
    p_value: The p_value associated with the chi-square statistic.
  """

  n = len(sample)
  sample_variance = np.var(sample, ddof=1)  # Sample variance

  # Calculate chi-square statistic
  chi_square_stat = (n - 1) * sample_variance / population_variance

  # Calculate degrees of freedom
  df = n - 1

  # Calculate p-value (two-tailed test)
  p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

  print("Chi-square Statistic:", chi_square_stat)
  print("p-value:", p_value)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence that the population variance is different from the hypothesized value.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence that the population variance is different from the hypothesized value.")

  return chi_square_stat, p_value

# Example usage:
sample_data = [25, 30, 28, 32, 27, 26, 29]
hypothesized_variance = 10

chi_square_stat, p_value = chi_square_test_for_variance(sample_data, hypothesized_variance)
```

Q25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
from statsmodels.stats.proportion import proportions_ztest

def z_test_for_proportions(count1, nobs1, count2, nobs2, alternative='two-sided', alpha=0.05):
  """
  Performs a two-sample Z-test for comparing proportions.

  Args:
    count1: Number of successes in the first group.
    nobs1: Total number of observations in the first group.
    count2: Number of successes in the second group.
    nobs2: Total number of observations in the second group.
    alternative: 'two-sided' (default), 'larger', or 'smaller'.
    alpha: Significance level (default: 0.05).

  Returns:
    z_stat: The calculated Z-statistic.
    p_value: The p-value.
  """

  z_stat, p_value = proportions_ztest([count1, count2], [nobs1, nobs2], alternative=alternative)

  print("Z-statistic:", z_stat)
  print("p-value:", p_value)

  if p_value < alpha:
    if alternative == 'two-sided':
      print("Reject the null hypothesis. There is evidence of a significant difference in proportions between the two groups.")
    elif alternative == 'larger':
      print("Reject the null hypothesis. There is evidence that the proportion in the first group is significantly larger than the proportion in the second group.")
    elif alternative == 'smaller':
      print("Reject the null hypothesis. There is evidence that the proportion in the first group is significantly smaller than the proportion in the second group.")
  else:
    print("Fail to reject the null hypothesis. There is no evidence of a significant difference in proportions between the two groups.")

  return z_stat, p_value

# Example usage:
count1 = 50  # Number of successes in group 1
nobs1 = 100  # Total number of observations in group 1
count2 = 40  # Number of successes in group 2
nobs2 = 80  # Total number of observations in group 2

z_stat, p_value = z_test_for_proportions(count1, nobs1, count2, nobs2)
```

Q26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_for_variance(sample1, sample2, alpha=0.05):
  """
  Performs an F-test to compare the variances of two samples.

  Args:
    sample1: The first sample data.
    sample2: The second sample data.
    alpha: Significance level (default: 0.05).

  Returns:
    f_statistic: The calculated F-statistic.
    p_value: The p-value associated with the F-statistic.
  """

  # Calculate sample variances
  var1 = np.var(sample1, ddof=1)  # Sample variance
  var2 = np.var(sample2, ddof=1)  # Sample variance

  # Determine which sample has larger variance
  if var1 > var2:
    f_statistic = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1
  else:
    f_statistic = var2 / var1
    df1 = len(sample2) - 1
    df2 = len(sample1) - 1

  # Calculate p-value
  p_value = 2 * (1 - f.cdf(f_statistic, df1, df2))  # Two-tailed test

  # Create box plots for visualization
  plt.figure(figsize=(8, 5))
  sns.boxplot(data=[sample1, sample2])
  plt.xlabel("Groups")
  plt.ylabel("Values")
  plt.title("Boxplot of Groups")
  plt.show()

  # Print results
  print("F-statistic:", f_statistic)
  print("p-value:", p_value)

  if p_value < alpha:
    print(f"Reject the null hypothesis at {alpha} significance level.")
    print("There is evidence that the variances of the two populations are significantly different.")
  else:
    print(f"Fail to reject the null hypothesis at {alpha} significance level.")
    print("There is no evidence that the variances of the two populations are significantly different.")

  return f_statistic, p_value

# Example usage:
sample1 = [25, 30, 28, 32, 27, 26, 29]
sample2 = [35, 40, 38, 42, 37, 36, 39]

f_statistic, p_value = f_test_for_variance(sample1, sample2)
```
